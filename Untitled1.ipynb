```python
from keras.models import Sequential 
from keras.layers import Dense

import numpy
from keras import regularizers
"""

    How to load data.
    How to define neural network in Keras.
    How to compile a Keras model using the efficient numerical backend.
    How to train a model on data.
    How to evaluate a model on data.

"""
#fixed random seed for reproducibility
numpy.random.seed(7)

#load data 
dataset = numpy.loadtxt("/home/kunaldargan/pima-indians-diabetes.csv",delimiter = ",")
#X and  Y

X = dataset[:,0:8]
Y = dataset[:,8]

#create model
model = Sequential()
model.add(Dense(12,input_dim = 8,activation = "relu",bias_regularizer = regularizers.l1(0.3)))
model.add(Dense(12, activation = "relu"))
model.add(Dense(12, activation = "sigmoid"))
model.add(Dense(12, activation = "relu"))
model.add(Dense(1, activation = "sigmoid",activity_regularizer = regularizers.l1(0.01)))

#compile the model
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])

#fit the model
model.fit(X, Y, epochs = 500, batch_size = 20)

#evalute the model
#scores = model.evaluate(X,Y)
#print("\n %s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

predictions = model.predict(X)
rounded = [ round(x[0]) for x in predictions ]
print(rounded)

model.save("/home/kunaldargan/model_kerasNN")
```
